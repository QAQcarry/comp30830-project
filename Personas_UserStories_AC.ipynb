# Persona: Tourist

Name: Ethan
Age: 30
Occupation: Tourist / Short-term Visitor
Technical Skill Level: Moderate

**Background**:

- Ethan is visiting Dublin for a short holiday.

- He enjoys exploring cities using convenient and affordable transportation options.

- He frequently uses navigation apps such as Google Maps when travelling.

- He is unfamiliar with the Dublin bike sharing system and station network.

**Goals**
- Find nearby bike stations quickly
- Check real-time availability of bikes and docking spcaes
- Travel conveniently between tourist arractions
- Avoid complicated sign-up and payment procedures

**Key needs**
- A simple and easy-to-use interface.
- Clear map-based visualization of bike stations.
- Real-time updates about bike and dock availability.
- Clicking on the site will display "Can you borrow/return now?" ,status indication
- Weather information integrated into the system.

**Pain points**
- Arriving at a station with no bikes available
- Not finding empty docks to return bikes
- Difficulty understading station locations.
- Inaccurate station status

**Usage Scenario**
- Ethan wants to travel from Radision Hotel to Temple Bar
- He opens the application to locate nearby bike stations
- he check the station status before his trip
- he confirms docking availability near his destination
- he reviews weather conditions before bike trip



# User Stories

**US1**

- As a tourist,
I want to locate nearby bike stations on a map,
So that I can quickly find available bikes and plan my trip.

**US2**
- As a tourist,
I want to check real-time bike and docking availability at each station,
So that I can avoid arriving at a station with no bikes or empty docks.

**US3**
- As a tourist,
I want to click on a station to see whether I can borrow or return a bike,
So that I can decide which station to use.

**US4**
- As a tourist,
I want to see weather information before travelling,
So that I can decide whether cycling is suitable.

**US5**
- As a tourist,
I want to see alternative nearby stations when my selected station is full or empty,
So that I can continue my journey without delays.

# Acceptance Criteria 

**View nearby bike stations on map**

- Given the user opens the application homepage,

- When the system loads the map view,

Then nearby bike stations are displayed clearly on an interactive map.

**Check real-time bike and dock availability**
- Given the user is viewing a bike station on the map,

- When the user selects a specific station,

- Then the system displays the real-time number of available bikes and empty docking spaces.

**Identify whether a bike can be borrowed or returned**
- Given the user selects a bike station,

- When the station availability information is displayed,

- Then the system clearly indicates whether the user can borrow or return a bike at that station 
View station details before starting a trip

**View station details before starting a trip**
- Given the user plans a bike trip

- When the user checks the station details,

- Then the system shows station name, location, bike availability, and dock availability in a single view.

**Check docking availability near destination**

- Given the user selects a destination area,

- When the user searches for nearby bike stations,

- Then the system displays stations near the destination with available docking spaces.

**View weather conditions before travelling**

- Given the user is planning a bike journey,

- When the user views the trip information,

- Then the system displays current weather conditions relevant to cycling.

**Access the system without complex sign-up**

- Given the user opens the application for the first time,

- When the user navigates through the system,

- Then the user can access station and availability information without mandatory account registration.

**Receive accurate and updated station information**

- Given the system retrieves bike station data from external APIs,

- When the data is refreshed,

- Then the displayed station availability reflects the most recent and accurate information.
