Persona: Tourist

Name: Ethan
Age: 30
Occupation: Tourist / Short-term Visitor
Technical Skill Level: Moderate

**Background**:

- Ethan is visiting Dublin for a short holiday.

- He enjoys exploring cities using convenient and affordable transportation options.

- He frequently uses navigation apps such as Google Maps when travelling.

- He is unfamiliar with the Dublin bike sharing system and station network.

**Goals**
- Find nearby bike stations quickly
- Check real-time availability of bikes and docking spcaes
- Travel conveniently between tourist arractions
- Avoid complicated sign-up and payment procedures

**Key needs**
- A simple and easy-to-use interface.
- Clear map-based visualization of bike stations.
- Real-time updates about bike and dock availability.
- Clicking on the site will display "Can you borrow/return now?" ,status indication
- Weather information integrated into the system.

**Pain points**
- Arriving at a station with no bikes available
- Not finding empty docks to return bikes
- Difficulty understading station locations.
- Inaccurate station status

**Usage Scenario**
- Ethan wants to travel from Radision Hotel to Temple Bar
- He opens the application to locate nearby bike stations
- he check the station status before his trip
- he confirms docking availability near his destination
- he reviews weather conditions before bike trip



**User Stories**

**US1**

- As a tourist,
I want to locate nearby bike stations on a map,
So that I can quickly find available bikes and plan my trip.

**US2**
- As a tourist,
I want to check real-time bike and docking availability at each station,
So that I can avoid arriving at a station with no bikes or empty docks.

**US3**
- As a tourist,
I want to click on a station to see whether I can borrow or return a bike,
So that I can decide which station to use.

**US4**
- As a tourist,
I want to see weather information before travelling,
So that I can decide whether cycling is suitable.

**US5**
- As a tourist,
I want to see alternative nearby stations when my selected station is full or empty,
So that I can continue my journey without delays.

**Acceptance Criteria for User Story**

- The system shall display a map interface showing all available bike stations.

- The system shall display the geographic location of each station using markers.

- The system shall allow users to zoom and navigate the map.

- The system shall load station locations within an acceptable response time.
